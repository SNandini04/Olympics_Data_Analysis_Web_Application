```python
def fetch_medal_tally(years,country):
    flag = 0
    if years == "Overall" and country == "Overall":
        temp_df = medal_df
    if years =="Overall" and country != "Overall":
        flag = 1
        temp_df = medal_df[medal_df['country'] == country]
    if years !="Overall" and country == "Overall":
        temp_df = medal_df[medal_df['years'] == int(years)] 
    if years !="Overall" and country != "Overall":
        temp_df = medal_df[(medal_df['years']== years) & (medal_df['country']== country)]

    if flag == 1:
        x = temp_df.groupby('years').sum()[['Gold','Silver','Bronze']].sort_values('years').reset_index()
    else:
        x = temp_df.groupby('country').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()

    x['total'] = x['Gold'] + x['Silver'] + x['Bronze']
    print(x)
```
